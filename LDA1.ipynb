```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import PCA
from sklearn import datasets

# 1.load data sets
iris = datasets.load_iris()
x = iris.data
y = iris.target
data = pd.DataFrame(data=np.c_[iris['data'], iris['target']],columns=iris['feature_names'] + ['target'])
print(data.head())

x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values

# 2.standardising data
scaler=StandardScaler()
x_std=scaler.fit_transform(x)

#LDA
lda=LDA(n_components=2)
x_lda=lda.fit_transform(x_std,y)

#visualization of LDA transformed data
plt.figure(figsize=(8,6))
for label in np.unique(y):
  plt.scatter(x_lda[y==label,0],x_lda[y==label,1],label=label)
plt.title("LDA iris data projection")
plt.xlabel("LDA1")
plt.ylabel("LDA2")
plt.legend()
plt.show()

```

```python
# 5.PCA for comparision
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_std)
plt.figure(figsize=(8,6))
for label in np.unique(y):
  plt.scatter(x_pca[y==label,0],x_pca[y==label,1],label=label)
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("PCA iris data projection")
plt.legend()
plt.show()
```
